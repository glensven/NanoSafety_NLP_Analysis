```python
# Dependencies
import requests
import json
import pandas as pd
from config import api_key
from pandas.io.json import json_normalize  
```

## GET Scopus Author Information Using Single Scopus Author ID

```python
#https://dev.elsevier.com/documentation/AuthorSearchAPI.wadl
url = "http://api.elsevier.com/content/author?"


headers = {
     "X-ELS-APIKey": api_key,
     'Accept':'application/json'
}

parameters = {
    "author_id": "56503901200"
   
}

response = requests.get(url, headers=headers, params=parameters)

print(response.url)
print(response.status_code)

response_json = response.json()
print(json.dumps(response_json, indent=4, sort_keys=True))
```

## GET Scopus Article Information Using Single Scopus Author ID

```python
#https://kitchingroup.cheme.cmu.edu/blog/2015/04/03/Getting-data-from-the-Scopus-API/
#https://dev.elsevier.com/documentation/AuthorSearchAPI.wadl
#https://dev.elsevier.com/guides/ScopusSearchViews.htm

url = "http://api.elsevier.com/content/search/scopus?"

headers = {
     "X-ELS-APIKey": api_key,
     'Accept':'application/json'
}

fieldList = ["dc:title", "prism:publicationName", "dc:identifier", "eid", "prism:doi", "pubmed-id", "subtypeDescription"]
parameters = {
    "query": "AU-ID(56503901200)",
     "field": ",".join(fieldList),
    "suppressNavLinks" : "true"
}


response = requests.get(url, headers=headers, params=parameters)

print(response.url)
print(response.status_code)

response_json_2 = response.json()
results = json.dumps(response_json_2, indent=4, sort_keys=True)
print(results)
```

```python
new_response = response_json_2["search-results"]["entry"]
json_normalize(new_response)
```
